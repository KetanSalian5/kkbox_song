# Feature Engineering 

Seperating  genre_ids, Artist Name, Composer, Lyricit and counting them as in a single row there are more than one

```python
# Explane the genre_ids
song_data['genre_ids'].head(10) # the line 5 and 9 have 2 or more
```

```python
def counting_genre_ids_artist_composer_lyricist(data):
    count = []
    for ids in data:
        try:
            if len(ids) != 0:
                count_ids = 1
                for i in ids:
                    if i == '|':
                        count_ids+=1
                count.append(count_ids)
        except TypeError:
            count.append(0)
    return count

song_data['count_of_genre_ids'] = counting_genre_ids_artist_composer_lyricist(song_data['genre_ids'])
song_data['count_of_artist'] = counting_genre_ids_artist_composer_lyricist(song_data['artist_name'])
song_data['count_of_composer'] = counting_genre_ids_artist_composer_lyricist(song_data['composer'])
song_data['count_of_lyricist'] = counting_genre_ids_artist_composer_lyricist(song_data['lyricist'])

song_data.head(10)
```

*we will handle missing values in three ways or make three models and see which performs the best:-
* 1: We will use mode to replace values in the column
* 2: we will remove them from the dataset
* 3: We will add a new category as missing in the column

# 1.Filling the data with Mode

```python
# mode_merged_data.columns
```

```python
mode_merged_data = merged_music_data.copy()

for col in mode_merged_data.columns:
    mode_merged_data[col].fillna(mode_merged_data[col].mode()[0],inplace=True)
```

```python
mode_merged_data.isnull().sum()
```

```python
#replacing string variables with numeric values
#for that we have to use the Sklearn preprocessing method
```

```python
columns = ['user_id', 'song_id', 'source_system_tab', 'source_screen_name',
       'source_type', 'gender', 'song_length', 'genre_ids', 'artist_name',
       'composer', 'lyricist', 'language']
from sklearn import preprocessing
lable_encoder = preprocessing.LabelEncoder()

def lable_encoding (data):
    for col in columns:
        print(col)
        data[col] = lable_encoder.fit_transform(data[col])
        
    return(data)
lable_encoding(mode_merged_data)
```

```python
missing_label_merged_data = merged_music_data.copy()

for col in missing_label_merged_data.columns:
    if col != 'language':
        missing_label_merged_data[col].fillna('missing',inplace=True)
    else:
        missing_label_merged_data[col].fillna(0,inplace=True)
```

# Implementing the models
* We will be using AUC score for selecting the model as it metic which is used on kegge

```python
from sklearn.model_selection import train_test_split
import math
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score


x_train_mode, x_test_mode, y_train_mode, y_test_mode = train_test_split(mode_merged_data.drop(['target'],axis=1),
                                                                        mode_merged_data['target'],test_size=0.20,random_state=40)
x_train_removed_null, x_test_removed_null, y_train_removed_null, y_test_removed_null = train_test_split(removed_null_data.drop(['target'],
                                                                                                                                  axis=1),
                                                                                                           removed_null_data['target'],test_size=0.20,random_state=40)

x_train_missing_label, x_test_missing_label, y_train_missing_label, y_test_missing_label = train_test_split(missing_label_merged_data.drop(['target'],axis=1),
                 missing_label_merged_data['target'],test_size=0.20,random_state=40)
```
